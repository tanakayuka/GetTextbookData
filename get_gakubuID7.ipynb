```python
cource_title_list = []
for i in range(6):
    gcd = i + 1
    url = "http://www.ocw.titech.ac.jp/index.php?module=General&action=T0100&GakubuCD={}&lang=JA".format(gcd)
    html = requests.get(url)
    soup = bs4.BeautifulSoup(html.content, "html.parser")
    titles = soup.find_all("td", class_ ="course_title")
    print("学部CD{}: 科目数{}".format(gcd,len(titles)))
    for raw_title in titles:
        title = str(raw_title).split("\n")
        cource_title_list.append(title[2].replace(" ",""))
        
    sleep(15)
    
```
